```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

## 텍스트 : 훈민정음 언해본 서문

```python
text = open(r'C:\Users\user\Desktop\30102\훈민정음언해본.txt').read()
text
```

```python
img = r'C:\Users\user\Desktop\30102\세종대왕.jpg'
Image.open(img)
```

```python
Ip_mask = np.array(Image.open(img))
Ip_mask
```

```python
Ip_wc = WordCloud(    
    font_path = r'‪C:\Windows\Fonts\batang.ttc',
    background_color = 'white',
    mask=Ip_mask,
    max_font_size=50,
    contour_width=5,
    contour_color='skyblue'
).generate(text)
```

```python
Ip_wc.words_
Ip_wc.to_image()
```
