```python
t = np.arange(0,12,0.5)
y1 = np.sin(t)
y2 = np.cos(t)
plt.grid()
plt.plot(t,y1,'y--',marker='o')
plt.plot(t,y2,'r--',marker='*')
plt.title('SinCos')
plt.xlabel('Y1')
plt.ylabel('Y2')
plt.legend(['sin','cos'])
```

```python
x = np.arange(0,5,1)
y1 = x+1
y2 = x+2
y3 = x+3
y4 = x+4

plt.subplot(1,2,1)
plt.plot(x,y1,'ro-',color='skyblue')
plt.plot(x,y2,'r--',color='pink')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y1','y2'])

plt.subplot(1,2,2)
plt.plot(x,y3,'g^-',color='blue')
plt.plot(x,y4,'r*-',color='hotpink')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y3','y4'])
```
